# Replenishment Simulation Demo

This notebook shows how to run the replenishment simulation with a historical-error safety stock policy.

```python
from janrth.replenishment_sim import HistoricalErrorSafetyStockPolicy, ReplenishmentSimulation, SimulationConfig

demand = [12, 9, 11, 10, 13, 8, 9, 12, 10, 11]
config = SimulationConfig(
    initial_stock=20,
    lead_time_days=2,
    time_step_days=1,
    holding_cost_per_unit_per_day=0.1,
    stockout_cost_per_unit_per_day=1.0,
    forecast_window=3,
)
policy = HistoricalErrorSafetyStockPolicy(safety_factor=1.5)
simulation = ReplenishmentSimulation(config=config, policy=policy)
results = simulation.run(demand)
results[:5]
```
